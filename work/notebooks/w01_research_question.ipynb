## The question: decision, action, cost of a wrong call

Research Question:
Which pages should be prioritized for content refresh based on their current performance and freshness?

Decision:
Determine whether a page should be refreshed now or later.

Action:
The SEO or content team updates the selected pages with improved and relevant content.

Cost of a Wrong Call:
Refreshing the wrong pages wastes time and resources, while ignoring important pages may result in continued traffic loss.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

declining = (df["trend_direction"].str.lower() == "down").sum()
improving = (df["trend_direction"].str.lower() == "up").sum()

print("Total Pages:", len(df))
print("Pages with Declining Trend:", declining)
print("Pages with Improving Trend:", improving)
```
